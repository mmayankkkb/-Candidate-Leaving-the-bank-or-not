```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
ds=pd.read_csv("Churn_Modelling.csv")
ds.head()
```

```python
X=ds.iloc[:,3:13]
y=ds.iloc[:,13]
X
```

```python
X['Gender']=np.where(X['Gender']=='Female',0,1)
X
```

```python
y
```

```python
# from sklearn.preprocessing import OneHotEncoder
# Oh=OneHotEncoder()

temp=pd.get_dummies(ds['Geography'],drop_first=True)
X=pd.concat([X,temp],axis=1)
X
```

```python
X.drop('Geography',inplace=True,axis=1)
X
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(X,y,random_state=30,test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
```

```python
train_X=Sc.fit_transform(train_X)
test_X=Sc.fit_transform(test_X)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,max_depth=20,max_leaf_nodes=10,max_features='auto')
rf.fit(train_X,train_y)
```

```python
RFans=rf.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score
RFScore =accuracy_score(RFans,test_y)*100
RFScore
```

```python
#ANN are build with fully connected layers

from tensorflow.keras.models import Sequential #class
from keras.layers import Dense # class
```

```python
classifier= Sequential()
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=11)) #units =last layers neuron#
# this will add an input layer and a hidden layer
```

```python
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu')) # one more hidden layer
```

```python
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid')) # Output layer
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])    #compiling the network

#adam -> performs stoicastic gradeint descent
#loss= used in training  
# binary_crossentropy->for binary classification it calculates the loss predicted labeled and true label
# metrices= for evaluation the network
# accuracy-> total and count(that are used to compute the frequency with y_pred and y_true)
```

```python
classifier.fit(train_X,train_y,batch_size=10,epochs=100)  

#epoch ->no of times one forward_feed and one backward pass of all the training ex.
#  or no. of times the algorithm will visit the dataset
```

```python
y_pred=classifier.predict(test_X)
y_pred
```

```python
y_pred=(y_pred>0.5)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
CMann=confusion_matrix(test_y,y_pred)
Acann=accuracy_score(test_y,y_pred)
```

```python
print(CMann)   #     [0,1]-> false predicted incorrect ,  [1,0]-> true predicted incorrect
```

```python
print(Acann*100)
```
